```python
fig, ax = plt.subplots()
plt.axvline(x = np.mean(mydata.Discussion), color='orange')
plt.axvline(x = np.median(mydata.Discussion), color='green')
_ = plt.hist(mydata.Discussion, color = 'lightgray')
```

```python
mydata.Discussion.var()
```

```python
sigma = mydata.Discussion.std()
sigma
```

```python
xbar = mydata.Discussion.mean()
```

```python
fig, ax = plt.subplots()
plt.axvline(x = xbar, color='orange')
plt.axvline(x = xbar+sigma, color='olivedrab')
plt.axvline(x = xbar-sigma, color='olivedrab')
_ = plt.hist(mydata.Discussion, color = 'lightgray')
```
